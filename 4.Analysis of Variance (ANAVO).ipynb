```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Pre-PlacementDataset.csv")
df
```

```python
df.isnull().sum()
```

```python
import scipy.stats as stats
"ONEWAY ANAVO"
stats.f_oneway(df["ssc_p"],df["hsc_p"],df["degree_p"])
```

```python
"""Here the p-value is 0.5110602818995302 this is above 5% so H0 -null hypothesis accepted,
there is no difference between those features """
```

```python
mf=pd.DataFrame()
mf["ssc_p"]=df["ssc_p"]
mf["hsc_p"]=df["hsc_p"]
df["salary"]=df["salary"]
mf
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
model = ols('salary ~ C(ssc_p) + C(hsc_p) + C(ssc_p):C(hsc_p)', data=mf).fit()
sm.stats.anova_lm(model, typ=2)
```
